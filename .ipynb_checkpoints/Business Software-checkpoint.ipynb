## Functions

```python
def get_pageids(page):
    page_query = 'http://en.wikipedia.org/w/api.php?action=query&format=json&list=categorymembers&cmpageid={}&cmnamespace=14&cmtype=subcat'.format(page)
    page_id = requests.get(page_query).json()['query']['categorymembers']
    return page_id
```

```python
def get_sub_cat(lists):
    next_level_sub = []
    for pages in lists:
        next_level_sub.append(get_pageids(pages))
    return next_level_sub
```

```python
def clean_list(ids):
    list_cleaner = []
    for a in range(len(ids)):
        for b in ids[a]:
            list_cleaner.append(b)
    lists_cleaned = pd.DataFrame(list_cleaner)
    return lists_cleaned
```

```python
def get_pages_articles(lists3):
    all_pages_needed = []
    for number in lists3:
        pagequery = 'http://en.wikipedia.org/w/api.php?action=query&format=json&&list=categorymembers&cmpageid={}&cmlimit=max'.format(number)
        search_page55 = requests.get(pagequery)
        pages = search_page55.json()['query']['categorymembers']
        all_pages_needed.append(pages)
    return all_pages_needed
```
